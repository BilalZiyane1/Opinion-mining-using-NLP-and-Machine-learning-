```python
import pickle
import os
from google.colab import drive
import pandas as pd
import numpy as np

# Mount Google Drive
drive.mount('/content/drive')

# Define the directory in your Google Drive where you saved the files
load_dir = '/content/drive/My Drive/sentiment_models/'

# Load the vectorizer
with open(os.path.join(load_dir, 'vectoriser-ngram-(1,2).pickle'), 'rb') as file:
    vectoriser = pickle.load(file)

# Load the models
with open(os.path.join(load_dir, 'Sentiment-BNB.pickle'), 'rb') as file:
    BNBmodel = pickle.load(file)

with open(os.path.join(load_dir, 'Sentiment-SVC.pickle'), 'rb') as file:
    SVCmodel = pickle.load(file)

with open(os.path.join(load_dir, 'Sentiment-LR.pickle'), 'rb') as file:
    LRmodel = pickle.load(file)

# Define some test sentences
test_sentences = [
    'I love this product!',
    'This is the worst product I have ever used.',
    'The quality of the item is okay, but it is a bit expensive.',
    'I am very disappointed with the service.',
    'The movie was amazing and I highly recommend it.'
]

# Transform the test sentences using the vectorizer
test_vectors = vectoriser.transform(test_sentences)

# Make predictions using each model
BNB_predictions = BNBmodel.predict(test_vectors)
SVC_predictions = SVCmodel.predict(test_vectors)
LR_predictions = LRmodel.predict(test_vectors)

# Print the predictions
print('Bernoulli Naive Bayes predictions:', BNB_predictions)
print('Support Vector Classifier predictions:', SVC_predictions)
print('Logistic Regression predictions:', LR_predictions)
```
